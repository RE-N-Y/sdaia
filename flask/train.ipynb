```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
from xgboost import XGBClassifier

X = pd.read_csv("fetal_health.csv")
X.describe()
```

```python
scaler = StandardScaler()
model = XGBClassifier()

X["fetal_health"] = X["fetal_health"].map({1:0, 2:1, 3:2})
y = X.pop("fetal_health")
```

```python
columns = X.columns.tolist()
```

```python
scaler.fit(X)
X = scaler.transform(X)
X.shape
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state=42)

model.fit(X_train, y_train)
```

```python
from sklearn.metrics import accuracy_score

accuracy_score(model.predict(X_test), y_test)
```

```python
import pickle as pkl

with open("pipeline.pkl", "wb") as f:
    pkl.dump({ "scaler":scaler, "model":model, "columns":columns }, f)
```

```python
import json

sample = { column:value for value, column in zip(X_test[42], columns) }

with open("sample.json", "w") as f: 
    json.dump(sample,f)
```
